# What are the most popular forms?

Matt has provided some stats from gov.uk. They show page visits for various pages where
there is a form. These should allow us to idenitify which are the most commonly completed 
forms. 

```python
import pandas as pd
from form_processor import get_all_files

%load_ext autoreload
%autoreload 2


form_stats = pd.read_csv("content-data-export-from-2023-08-05-to-2024-08-04-from-all-organisations-in-form.csv")
form_stats.head()
```

```python
from pathlib import Path

folder_path = "forms_scrape"
all_files = get_all_files(folder_path)
pdf_files = [Path(file) for file in all_files if str(file).lower().endswith(".pdf")]
```

```python
# Which forms should we process 

forms_to_process = [path for path in pdf_files if str(path).split("/")[1] in form_stats.head(100)['Title'].str.replace(" ", "_").to_list() ]
len(forms_to_process)
```
